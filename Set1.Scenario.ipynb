```python
# 3. Write logic to determine the bonus amount based on the given salary.

emp_salary = int(input('Enter your salary :'))
if (emp_salary > 50000):
    bonus = emp_salary + (emp_salary*0.1)
    print('Your bonus amount is :', bonus)
else:
    print('No bonus for you')
```

```python
# 5. Write logic to take a word or sentence as input and produce its reversed version.

string = input('Enter a string :')
reverse_string = ''
for char in string:
    reverse_string = char + reverse_string
print('Reverse string is :', reverse_string)

text = input('Enter a string :')
reverse_text = text[::-1]
print(reverse_text)
```

```python
# 8. Write logic to check if a user has enough balance before allowing a withdrawal and update the remaining balance accordingly.

customer_balance = {'Raji' : 10000,
                    'Rubhini' : 18000,
                    'Ramisha' : 12000}
Name = input('Enter your Name: ')
Withdraw = int(input('Enter the Amount to be withdrawn: '))

if (customer_balance[Name] < Withdraw):
    print('Insufficient Balance')
else:
    customer_balance[Name] = customer_balance[Name] - Withdraw 
    print('Collect your cash! Current Account Balance :', customer_balance[Name])
               
```

```python
# 9. Write logic to determine whether a given year is a leap year

Year = int(input('Enter the Year: '))
if (Year % 4 == 0 and Year % 100 != 0) or (Year % 400 == 0):
    print(Year, 'is a Leap year')
else:
    print(Year, 'is not a Leap year')
           
```
